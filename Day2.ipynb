```python
import pprint
```

```python
qry=[{"$group":{"_id":"$year","max_runtime":{"$max":"$runtime"}}}]
for x in mov.aggregate(qry): # 1 is for display 0 is for not to display
    pprint.pprint(x)
```

```python
# displaying 4th and 5th document
for x in mov.find().limit(2).skip(3): # 1 is for display 0 is for not to display
    print(x)
```

```python
# actor: Dennis Hopper movies
for x in mov.find({'actors':'Dennis Hopper'},{'_id':0,'title':1}): # 1 is for display 0 is for not to display
    pprint.pprint(x)
```

```python
for x in mov.distinct('year'): # 1 is for display 0 is for not to display
    pprint.pprint(x)
```

```python
for x in mov.aggregate([{"$group":{"_id":"$year","count":{"$sum":1}}}]):
    pprint.pprint(x)
```

```python
# retrive only votes from imdb, suppose 'imdb': {'id': 'tt0076759', 'rating': 8.7, 'votes': 822849.0}
for x in mov.find({},{'_id':0,'imdb.votes':1}):
    print(x)
```

```python
# movies name after 1990
for x in mov.find({'year':{'$gt':1990}}, {'title':1,'_id':0}):
    print(x)
```
