```python
import nltk
from gensim.models import Word2Vec
from nltk.corpus import stopwords
import re
```

```python
#processing data
text = re.sub(r'\[0-9]*\]',' ' ,paragraph)
text = re.sub(r'\s+',' ',text)
text = text.lower()
text = re.sub(r'\d',' ',text)
text = re.sub(r'\s+',' ',text)
stop_updated =  ["...","..","!!"]
```

```python
#preparing data set
sentences = nltk.sent_tokenize(text)
```

```python
sentences[:5]
```

```python
sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
```

```python
for i in range(len(sentences)):
    sentences[i]= [word for word in sentences[i] if word not in (stopwords.words('english') and stop_updated)]
```

```python
sentences
```

```python
#training the word2Vec model
model = Word2Vec(sentences, min_count=1)
```

```python
model
```

```python
words= model.wv.vocab
words
```

```python
#Finding vector words
vector = model.wv['terms'] # var is a word present in test
vector
```

```python
# just similar words
similar = model.wv.most_similar('controls')
```
