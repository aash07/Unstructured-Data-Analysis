#### Load Google's Pretrained model

```python
google_model ='GoogleNews-vectors-negative300.bin.gz'
embeddings = gensim.models.KeyedVectors.load_word2vec_format(google_model, binary=True)
```

```python
embeddings
```

```python
embeddings['google']
```

```python
embeddings['google'].shape
```

### create document vectors using word embeddings

```python
docs_vectors = pd.DataFrame()
for doc in df.clean_review:
    temp = pd.DataFrame()
    words = doc.split(' ')
    for word in words:
        try:
            word2vec = embeddings[word] #gives vector of length
            temp= temp.append(pd.Series(word2vec), ignore_index=True)
        except:
            pass
    doc_vector = temp.mean()
    docs_vectors = docs_vectors.append(doc_vector, ignore_index=True)
docs_vectors.head()
```

- check for na n drop those
- build a classification model
- rebuild the model by not droping na - use ffill
- docs_vectors = docs_vectors.fillna(metheod='ffill')

```python
docs_vectors.isna().sum()
```

```python
docs_vectors = docs_vectors.fillna(method='ffill')
```

```python
x = docs_vectors
y = df.sentiment.values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
```

```python
from sklearn.metrics import accuracy_score
```

```python
ab= AdaBoostClassifier(n_estimators=1000)
ab.fit(x_train, y_train)
y_pred = ab.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
gb= GradientBoostingClassifier(n_estimators=1000)

gb.fit(x_train, y_train)
y_pred3 = gb.predict(x_test)
accuracy_score(y_test, y_pred3)
```

```python
from xgboost import XGBClassifier
```

```python
xgb = XGBClassifier()

xgb.fit(x_train, y_train)
y_pred4 = xgb.predict(x_test)
accuracy_score(y_test, y_pred4)
```
